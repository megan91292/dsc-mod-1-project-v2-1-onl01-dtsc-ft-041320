```python
import ast
import pandas as pd
```

```python
df= pd.read_csv('cleandata/Revenue_Budget_Genre.csv', index_col = 0)
```

```python
df.head()
```

```python
df.dropna(inplace = True)
```

```python
df.drop(df.loc[df['genres']=='[]'].index, inplace = True)
```

```python
df['genres'] = df['genres'].apply(ast.literal_eval)
```

# Using random.choice to randomly allocated each movie only one genre from the array of genres they are under

```python
import random
```

```python
df['randomgenre'] = df['genres'].apply(lambda x: random.choice(x))
```
